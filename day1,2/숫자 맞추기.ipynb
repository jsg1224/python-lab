```python
import random

answer_number = random.randint(1,100)
answer_number
count = 0
print("숫자 맞추기 게임에 오신 것을 환영합니다!")
print("컴퓨터가 1부터 100 사이의 숫자를 선택했습니다")
print("그 숫자를 맞춰보세요!")

while True:
    guess = int(input("숫자를 입력하세요 (1~100): "))
    count += 1


    if answer_number == guess:
        print(f'축하합니다! 정답은 {answer_number}였습니다!')
        print(f'총 {count}번 만에 맞췄습니다')
        break

    elif answer_number > guess:
        print('틀렸습니다. 더 큰 숫자입니다')
    else:
        print('틀렸습니다. 더 작은 숫자입니다')
```
